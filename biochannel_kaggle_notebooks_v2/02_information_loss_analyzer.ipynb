```
!pip install -q 'kagglehub[pandas-datasets]'

from pathlib import Path

import kagglehub
import pandas as pd

HANDLE = "crawford/gene-expression"
root = Path(kagglehub.dataset_download(HANDLE))
full = pd.read_csv(root / "data_set_ALL_AML_train.csv")
patient_cols = [c for c in full.columns if str(c).isdigit()]
# Rows = patients, columns = first genes (signal vs output channels for MI demo)
Xt = full[patient_cols].astype(float).T.reset_index(drop=True)
df = Xt.iloc[:, :3]
df.head()
```

```

import numpy as np
from sklearn.metrics import mutual_info_score

signal = df.iloc[:,1]
output = df.iloc[:,2]

signal_d = pd.cut(signal, bins=10, labels=False)
output_d = pd.cut(output, bins=10, labels=False)

mi = mutual_info_score(signal_d, output_d)
mi
```
