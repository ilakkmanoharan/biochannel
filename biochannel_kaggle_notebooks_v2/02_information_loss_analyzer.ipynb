```

import numpy as np
from sklearn.metrics import mutual_info_score

signal = df.iloc[:,1]
output = df.iloc[:,2]

signal_d = pd.cut(signal, bins=10, labels=False)
output_d = pd.cut(output, bins=10, labels=False)

mi = mutual_info_score(signal_d, output_d)
mi
```
