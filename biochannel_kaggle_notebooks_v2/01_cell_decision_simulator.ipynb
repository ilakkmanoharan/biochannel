# Cell Decision Simulator
Download + preprocess RNA-seq dataset

```
# Source: UCI ML Repository — Gene expression cancer RNA-Seq (PANCAN HiSeq)
# Dataset page: https://archive.ics.uci.edu/dataset/401/gene+expression+cancer+rna+seq
# The tarball below matches the Kaggle/UCI `data.csv` + `labels.csv` layout (801 × 20k genes).
```

```
import tarfile
import urllib.request
from pathlib import Path

import pandas as pd

WORKDIR = Path("/kaggle/working") if Path("/kaggle/working").exists() else Path.cwd() / "_biochannel_data"
WORKDIR.mkdir(parents=True, exist_ok=True)

RNA_SEQ_TAR_URL = (
    "https://archive.ics.uci.edu/ml/machine-learning-databases/00401/"
    "TCGA-PANCAN-HiSeq-801x20531.tar.gz"
)
SUBDIR = WORKDIR / "TCGA-PANCAN-HiSeq-801x20531"

if not (SUBDIR / "data.csv").exists():
    tar_path = WORKDIR / "TCGA-PANCAN-HiSeq-801x20531.tar.gz"
    if not tar_path.exists():
        urllib.request.urlretrieve(RNA_SEQ_TAR_URL, tar_path)
    with tarfile.open(tar_path, "r:gz") as tf:
        tf.extractall(WORKDIR)

expr = pd.read_csv(SUBDIR / "data.csv", index_col=0)
labels = pd.read_csv(SUBDIR / "labels.csv", index_col=0)
df = expr.join(labels).rename(columns={"Class": "label"})
df.head()
```

```

# Preprocess
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

X = df.drop(columns=['label'], errors='ignore')
X = np.log1p(X)
X_scaled = StandardScaler().fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
X_pca[:5]
```
