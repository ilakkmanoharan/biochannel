```
!pip install -q 'kagglehub[pandas-datasets]'

from pathlib import Path

import kagglehub
import pandas as pd

HANDLE = "crawford/gene-expression"
root = Path(kagglehub.dataset_download(HANDLE))
full = pd.read_csv(root / "data_set_ALL_AML_train.csv")
patient_cols = [c for c in full.columns if str(c).isdigit()]
df = full[patient_cols].astype(float).T.reset_index(drop=True)
df.head()
```

```
# Lightweight processing
import numpy as np

X = np.log1p(df.select_dtypes(include=[float]))
pd.DataFrame(X).head()
```
