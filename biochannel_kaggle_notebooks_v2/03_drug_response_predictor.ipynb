```
# Source: Genomics of Drug Sensitivity in Cancer (GDSC) — aggregated IC50 table on Kaggle.
# Handle: `samiraalipour/genomics-of-drug-sensitivity-in-cancer-gdsc` via `kagglehub`.
```

```
!pip install -q 'kagglehub[pandas-datasets]'

from pathlib import Path

import kagglehub
import pandas as pd

HANDLE = "samiraalipour/genomics-of-drug-sensitivity-in-cancer-gdsc"
root = Path(kagglehub.dataset_download(HANDLE))
path = root / "GDSC2-dataset.csv"
if not path.exists():
    path = root / "GDSC_DATASET.csv"

raw = pd.read_csv(path, low_memory=False)
df = raw[["CELL_LINE_NAME", "DRUG_NAME", "LN_IC50"]].dropna()
df.head()
```

```

from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder

le1, le2 = LabelEncoder(), LabelEncoder()
df['cell'] = le1.fit_transform(df.iloc[:,0])
df['drug'] = le2.fit_transform(df.iloc[:,1])

X = df[['cell','drug']]
y = df.iloc[:,2]

model = RandomForestRegressor()
model.fit(X,y)
model.predict(X[:5])
```
